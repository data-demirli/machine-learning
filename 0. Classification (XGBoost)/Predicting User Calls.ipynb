```python
"""
Function generates k stratified training & test sets and applies over/under sampling to the training sets only.
This enables cross validation without over/under sampling of the test sets.
Option to scale values for parametric models.
The first three items returned are passed into BayesSearchCV.
"""

def kfolds_func(df_x, df_y, ovsamp_strat, unsamp_strat, n_splits=5, scale_values=False):

    # Generating stratified k-fold indeces
    skf = StratifiedKFold(n_splits=n_splits, shuffle=True, random_state=1)
    skf_index = skf.split(df_x, df_y)
    
    df_train_x, df_train_y, df_test_x, df_test_y = {}, {}, {}, {}

    # Storing data sets in dictionaries, based on skf indeces, with dynamic key naming
    for i, (train_index, test_index) in enumerate(skf.split(df_x, df_y)):
        df_train_x["{0}".format(i)] = df_x.iloc[train_index]
        df_train_y["{0}".format(i)] = df_y.iloc[train_index]
        df_test_x["{0}".format(i)] = df_x.iloc[test_index]
        df_test_y["{0}".format(i)] = df_y.iloc[test_index]
    
    oversample = SMOTE(sampling_strategy=ovsamp_strat, random_state=1)
    undersample = RandomUnderSampler(sampling_strategy=unsamp_strat, random_state=1)
    df_x_final = pd.DataFrame()
    df_y_final = pd.DataFrame()
    index_tuples = []
    floor = 0
    ceiling = 0

    for k in df_train_x.keys():
        
        # Applies scaling if option enabled
        if scale_values:
            scaler = MinMaxScaler()
            scaler.fit(df_train_x[k])
            df_train_x[k] = pd.DataFrame(scaler.transform(df_train_x[k]), columns = df_train_x[k].columns)
            df_test_x[k] = pd.DataFrame(scaler.transform(df_test_x[k]), columns = df_test_x[k].columns)
        
        # Applies over/under sampling to traning sets
        if ovsamp_strat != None and unsamp_strat != None:
            df_train_x[k], df_train_y[k] = oversample.fit_resample(df_train_x[k], df_train_y[k])
            df_train_x[k], df_train_y[k] = undersample.fit_resample(df_train_x[k], df_train_y[k])
        elif ovsamp_strat != None:
            df_train_x[k], df_train_y[k] = oversample.fit_resample(df_train_x[k], df_train_y[k])
        elif unsamp_strat != None:
            df_train_x[k], df_train_y[k] = undersample.fit_resample(df_train_x[k], df_train_y[k])      
        
        # Tracks the indeces of training and test indeces for the final concatenated dataframe
        ceiling = floor+len(df_train_x[k])
        indeces_train = np.array(range(floor, ceiling))
        floor = ceiling
        ceiling = floor+len(df_test_x[k])
        indeces_test = np.array(range(floor, ceiling))
        floor = ceiling
        index_tuples.append((indeces_train, indeces_test))

        # Concatenates final training and test sets into a single dataframe
        df_x_final = pd.concat([df_x_final, df_train_x[k], df_test_x[k]], axis=0)
        df_y_final = pd.concat([df_y_final, df_train_y[k], df_test_y[k]], axis=0)
    
    
    # Resets index for the final concatenated dataset
    df_x_final = df_x_final.reset_index().drop(['index'], axis=1)
    df_y_final = df_y_final.reset_index().drop(['index'], axis=1)

    return df_x_final, df_y_final, index_tuples, df_train_x, df_train_y, df_test_x, df_test_y
```

```python
"""
Function applies over/under sampling on the entire dataset for final model training and predictions
"""

def sampling_func(df_x, df_y, ovsamp_strat, unsamp_strat):
  
    oversample = SMOTE(sampling_strategy=ovsamp_strat, random_state=1)
    undersample = RandomUnderSampler(sampling_strategy=unsamp_strat, random_state=1)
    
    # Applies over/under sampling to traning sets
    if ovsamp_strat != None and unsamp_strat != None:
        df_x, df_y = oversample.fit_resample(df_x, df_y)
        df_x, df_y = undersample.fit_resample(df_x, df_y)
    elif ovsamp_strat != None:
        df_x, df_y = oversample.fit_resample(df_x, df_y)
    elif unsamp_strat != None:
        df_x, df_y = undersample.fit_resample(df_x, df_y)
                                                                     
    df_x_final = df_x.reset_index().drop(['index'], axis=1)
    df_y_final = df_y.reset_index().drop(['index'], axis=1)

    return df_x_final, df_y_final
```

```python
"""
Data Preparation Function combines input datasets for pre-processing and model training:
 [1]Joins customer countries into main dataset (one-hot encoded)
 [2]Drops non-predictive columns
"""

def join_countries_func(df_x, df_countries=df_countries):

    df_joined = df_x.join(df_countries.set_index('mk_CurrentCustomer'), on='mk_CurrentCustomer', how='left')
    countries_ohe = pd.get_dummies(df_joined[['country']])

    df_customerIDs = df_joined[['mk_CurrentCustomer']]
    df_x_final = pd.concat([df_joined, countries_ohe], axis=1).drop(['mk_CurrentCustomer', 'ScoreDate', 'country'], axis=1)

    return df_x_final, df_customerIDs


"""
Feature Engineering Function generates two sets of new features based on existing features caputing weekly counts.
 [1] Aggregates: Weeks 1-3, 1-6, and 1-10
 [2] Ratios: Week 1 vs. sum(weeks 1-3), sum(weeks 1-6), and sum(weeks 1-10)
"""

def add_features_func(df_x):
    
    # The final 54 columns are excluded as they do not correspond to weekly count values
    target_columns = df_x.columns.values[0:-54]
    column_names = []
    
    df_feature_sums = pd.DataFrame()
    df_feature_ratios = pd.DataFrame()

    # Parsing the column names into their prefix/category
    for name in target_columns:
        column_names.append((name[::-1].split("_", 1)[1])[::-1])

    for prefix in set(column_names):
        # Generating first set of new features
        df_feature_sums[prefix+"_sum3w"] = df_x[prefix+"_g10"] + df_x[prefix+"_g9"]  + df_x[prefix+"_g8"]
        df_feature_sums[prefix+"_sum6w"] = df_x[prefix+"_g7"] + df_x[prefix+"_g6"]  + df_x[prefix+"_g5"] + df_feature_sums[prefix+"_sum3w"]
        df_feature_sums[prefix+"_sum10w"] = df_x[prefix+"_g4"] + df_x[prefix+"_g3"]  + df_x[prefix+"_g2"] + df_x[prefix+"_g1"] + df_feature_sums[prefix+"_sum6w"]
        
        # Generating second set of new features
        df_feature_ratios[prefix+"_ratio3w"] = df_x[prefix+"_g10"]/(0.01+df_feature_sums[prefix+"_sum3w"])
        df_feature_ratios[prefix+"_ratio6w"] = df_x[prefix+"_g10"]/(0.01+df_feature_sums[prefix+"_sum6w"])
        df_feature_ratios[prefix+"_ratio10w"] = df_x[prefix+"_g10"]/(0.01+df_feature_sums[prefix+"_sum10w"])
        
    return df_feature_sums, df_feature_ratios


"""
Data Preparation Function applies above two functions and final steps on input datasets
"""

def dataset_prep_func(df):  

    # Adding customer countries into main dataset and dropping non-predictive features
    df_x = df.drop(['target'], axis=1)
    df_x, _ = join_countries_func(df_x)

    # Generating two new sets of features from existing features 
    df_sums_x, df_ratios_x = add_features_func(df_x)
    # Removing game wins and margin ratios features; sums with negative values can lead to infinite ratios
    df_ratios_x = df_ratios_x.drop(['gw_ratio3w','gw_ratio6w','gw_ratio10w','mar_ratio3w','mar_ratio6w','mar_ratio10w'], axis=1)
    
    # Finalizaing attribute and target dataframes
    df_x_final = pd.concat([df_x, df_sums_x, df_ratios_x], axis=1)
    df_y_final = df[['target']]
    
    return df_x_final, df_y_final
```

```python
"""
Function returns selected features with importance values, resuting from logistic regression with l2 regularization
"""

def lr_feat_rank_func(df_x, df_y, C):
    
    # Fitting a logistic regression model to the data
    LogReg_model = LogisticRegression(penalty='l2', C=C, solver='liblinear', class_weight='balanced', random_state=1)
    LogReg_model.fit(df_x, df_y['target'])
    
    # Ranking feature importance from the model output
    feature_importance_lr = pd.DataFrame(list(X.columns), index=list(abs(LogReg_model.coef_))).sort_index(ascending=False)
    feature_importance_lr = feature_importance_lr.reset_index().set_axis(['Feature_Importance', 'Feature'], axis='columns')
    
    return feature_importance_lr
```

## Section 5 - Hyperparameter Tuning & Feature Selection

1. Training models with different sets of hyperparameters and features
2. Plotting results for each hyperparameter

```python
# Running XGBoost classification models
# Using a loop to select features (using cut-off points on the ranked set) 

results_df_xg = pd.DataFrame()
ov_samp, un_samp = (None, 0.2)
# Using a generator to store the set of features with different cut-off points
feature_sets_generator = (feature_importance_lr['Feature'][:x] for x in [-50, -100, -150])

search_space = {
    'n_estimators': Integer(80, 200),
#    'max_depth': Integer(2, 30),
    'learning_rate': Real(0.1, 1.0, prior='log-uniform'),
    'subsample': Real(0.5, 1.0),
#    'colsample_bytree': Real(0.5, 1.0),
#    'colsample_bylevel': Real(0.5, 1.0),
    'reg_lambda': Real(0.0, 50.0),
    'gamma': Real(0.0, 50.0)
}

for select_features in feature_sets_generator:
    #Generating the k-fold training & test sets (with over/under sampling applied) and indeces for use in BayesSearchCV
    df_x_final, df_y_final, index_tuples, _, _, _, _ = kfolds_func(X[select_features], Y, ov_samp, un_samp)

    XGB_model = XGBClassifier(random_state=1)
    BSCV_xg = BayesSearchCV(XGB_model, search_space, n_iter=10, cv=iter(index_tuples), scoring='roc_auc')
    BSCV_xg.fit(df_x_final, df_y_final['target'])
    
    # Combining the results tables of previous runs with different sets of features used
    results_df_xg = pd.concat([results_df_xg, pd.DataFrame(BSCV_xg.cv_results_).assign(Feature_Count=len(select_features))], axis=0)

# Storing the best hyperparameter values and the ranked features cut-off point
results_df_xg = results_df_xg.sort_values('mean_test_score', ascending=False).reset_index()
best_params_xg = results_df_xg.iloc[0]['params']
best_feature_count_xg = results_df_xg.iloc[0]['Feature_Count']

display(results_df_xg[0:3])
```

```python
# Plotting the trial results of each hyperparameter (colour coding refering to the number of features used)

sns.set_theme(rc={'axes.facecolor':'whitesmoke', 'axes.edgecolor': 'black'}, font_scale=1)
fig, axs = plt.subplots(3, (len(search_space)+1)//3, figsize=(15, 12), sharey=True)
plt.subplots_adjust(wspace=0.05, hspace=0.4)

for i, (param_category, ax) in enumerate(zip(search_space, axs.ravel())):
    sns.scatterplot(x='param_'+param_category, y='mean_test_score', hue='Feature_Count', data=results_df_xg, palette='colorblind',  ax=ax)
    ax.legend(facecolor='white', loc='lower right')
```

## Section 6 - Performance Metrics

Combining the results from 5 different cross-validation runs

```python
# Running 5-fold cross validation on the model using the best hyperparameters and storing the results

ov_samp, un_samp = (None, 0.2)
select_features_xg = feature_importance_lr['Feature'][:best_feature_count_xg]
_, _, _, df_train_x, df_train_y, df_test_x, df_test_y = kfolds_func(X[select_features_xg], Y, ov_samp, un_samp)
Y_pred_list, Y_pred_proba_list, Y_validate_list = [], [], []

# Looping through each of the 5 folds for cross validation
for k in df_train_x.keys():
    #Fitting a model with the best hyperparamters
    XGB_best_model = XGBClassifier(**best_params_xg, random_state=1)
    XGB_best_model.fit(df_train_x[k], df_train_y[k]['target'])
    
    # Storing predictions of each cross-validation in a list to later be concatenated
    Y_pred_list.append(XGB_best_model.predict(df_test_x[k]))
    Y_pred_proba_list.append(XGB_best_model.predict_proba(df_test_x[k]))
    Y_validate_list.append(df_test_y[k]['target'])

# Concatenating the predictions from each of the 5 runs
Y_pred = np.concatenate(Y_pred_list)
Y_pred_proba = np.concatenate(Y_pred_proba_list)
Y_validate = np.concatenate(Y_validate_list)
# Preparing ROC AUC scores on the combined results
roc_auc = roc_auc_score(Y_validate, Y_pred_proba[:,1])
fpr, tpr, _ = roc_curve(Y_validate, Y_pred_proba[:,1], pos_label=1)

print('{:.4}'.format(roc_auc), '*ROC AUC*')
```

```python
# Plotting the ROC Curve
# Results reflect combined predictions from 5 different runs

sns.set_theme(rc={'axes.facecolor':'whitesmoke', 'axes.edgecolor': 'black'}, font_scale=0.9)
fig, ax = plt.subplots()
ax.plot(fpr, tpr, label= 'ROC AUC: {:.1%}'.format(roc_auc))
ax.set_title("ROC Curves (5-Fold CV Average)", fontsize=13)
ax.set_xlabel("FPR")
ax.set_ylabel("TPR")
ax.legend(facecolor='white', loc='center right')
```

```python
# Setting a threshold to evaluate model performance at a specific point 
# Results reflect combined predictions from 5 different runs

threshold = 0.12
Y_pred_thresh = (np.array(Y_pred_proba[:,1]) >= threshold).astype(int)

# Confusion matrix values divided by 5, as the results are combined from 5 runs
conf_mat = confusion_matrix(Y_validate, Y_pred_thresh)//5
TPR = conf_mat[1,1]/(conf_mat[1,1]+conf_mat[1,0])
FPR = conf_mat[0,1]/(conf_mat[0,1]+conf_mat[0,0])
f1 = f1_score(Y_validate, Y_pred_thresh)

print('Metrics for threshold value of', '{:.1%}'.format(threshold), '\n\n', \
      'Model Confusion Matrix:', '\n', conf_mat, '\n\n', \
      '{:.1%}'.format(TPR), '*True Positive Rate*', '\n', \
      '{:.1%}'.format(FPR), '*False Positive Rate*', '\n', \
      '{:.3}'.format(TPR/FPR), '*TPR/FPR*', '\n', \
      '{:.1%}'.format(f1), '*F1-score*', '\n')
```
